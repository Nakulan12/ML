```python
my_list = [1, "hello", 3.14, True, 5, "world", 7]
print("Initialized list:", my_list)
```

```python
my_list.append("new_item")
print("List after appending 'new_item':", my_list)
```

```python
my_list.insert(2, "inserted_item")
print("inserted_item at index 2:", my_list)
```

```python
my_list.remove("inserted_item")
print(" 'inserted_item':", my_list)
```

```python
removed_item = my_list.pop(3)
print(f"List after removing item at index 3 ('{removed_item}'):", my_list)
```

```python
my_list = [str(item) for item in my_list]
my_list.sort()
print("List after converting to strings and sorting:", my_list)
```

```python
my_list.reverse()
print("List after reversing:", my_list)
```
